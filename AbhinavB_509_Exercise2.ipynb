```python
from nltk.tokenize import word_tokenize
g_tokens = word_tokenize(g_raw, language='greek')
print('Greek Tokens: ',g_tokens)

from nltk.tokenize import word_tokenize
f_tokens = word_tokenize(f_raw, language='french')
print('French Tokens: ',f_tokens)
```

```python
import jieba
c_tokens = jieba.cut(c_raw)
clist = list(c_tokens)
print('Chinese Tokens: ', clist)

from janome.tokenizer import Tokenizer
tokenizer = Tokenizer()
j_tokens = tokenizer.tokenize(j_raw)
japanese_tokens = []
for token in j_tokens:
    japanese_tokens.append(token.surface)
print("Japanese Tokens:", japanese_tokens)
```

```python
from indicnlp.tokenize import indic_tokenize
t_tokens = indic_tokenize.trivial_tokenize(t_raw)
print('Telugu tokens: ', t_tokens)

from indicnlp.tokenize import indic_tokenize
m_tokens = indic_tokenize.trivial_tokenize(m_raw)
print('Telugu tokens: ', m_tokens)
```
