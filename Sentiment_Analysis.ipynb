#Sentimental analysis or opinion mining is a technique in NLP to detect the input data is positive, Negative or Neutral.
##It emphasis on the emotional polarity of the text and detect the feelings such as happy, sad or Angry of a given text, such as a sentence, paragraph, or document.
##Sentiment analysis is commonly used to understand and quantify the subjective opinions, attitudes, and emotions expressed by individuals in text data, such as social media posts, customer reviews, news articles, and more.
##It can be applied to various domains, including business, marketing, social sciences, and customer service, to gain insights and make data-driven decisions.

#Polarity: Polarity refers to the emotion or sentiment expressed in a piece of text, whether it's positive, negative, or neutral. In sentiment analysis, polarity scores are often represented on a scale, where values close to +1 indicate strong positive sentiment, values close to -1 indicate strong negative sentiment, and values around 0 indicate neutral sentiment.

#Subjectivity: Subjectivity measures to what extent a piece of text expresses opinions, feelings, or beliefs rather than factual information. Texts that express personal opinions or emotions tend to be more subjective, while factual information tends to be more objective. Subjectivity scores can be represented on a scale as well, ranging from 0 to 1, where higher values indicate more subjective text and lower values indicate more objective text.

#Sentiment analysis techniques typically involve using machine learning algorithms or rule-based approaches to classify text into different sentiment categories, such as positive, negative, or neutral, based on features such as words, phrases, context, and sentiment lexicons.
##Machine learning algorithms used for sentiment analysis include supervised methods such as Naive Bayes, support vector machines, and deep learning models like recurrent neural networks (RNNs) and convolutional neural networks (CNNs).

#Sentiment analysis can also involve more advanced techniques, such as aspect-based sentiment analysis, which involves analyzing sentiment towards specific aspects or entities within a text, and emotion detection, which aims to identify specific emotions expressed in the text, such as joy, anger, or sadness.

#There are several tasks, including feature- or aspect-based sentiment analysis, polarity sentiment identification, and objective or subjective categorization.
##Words and phrases' subjectivity may vary depending on their context, and subjective sentences may appear in an objective document. Sentiment expressed toward particular aspects of entities is referred to as aspect-based sentiment analysis.
##Two factors that are utilised to score sentiment analysis are polarity and intensity. The sentiment's polarity shows whether it is negative, neutral, or positive. The sentiment's relative strength is indicated by intensity.

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense, Dropout, SpatialDropout1D
from tensorflow.keras.layers import Embedding

df = pd.read_csv("/content/drive/MyDrive/Datasets/Tweets.csv")
```

```python
tweet_df = df[['text','airline_sentiment']]
print(tweet_df.shape)
tweet_df.head(5)
```

```python
tweet_df = tweet_df[tweet_df['airline_sentiment'] != 'neutral']
print(tweet_df.shape)
tweet_df.head(5)
```

```python
tweet_df["airline_sentiment"].value_counts()
```

```python
sentiment_label = tweet_df.airline_sentiment.factorize()
sentiment_label
```

```python
tweet = tweet_df.text.values
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(tweet)
vocab_size = len(tokenizer.word_index) + 1
encoded_docs = tokenizer.texts_to_sequences(tweet)
padded_sequence = pad_sequences(encoded_docs, maxlen=200)
```

```python
print(tokenizer.word_index)
```

```python
print(tweet[0])
print(encoded_docs[0])
```

```python
print(padded_sequence[0])
```

```python
embedding_vector_length = 32
model = Sequential()
model.add(Embedding(vocab_size, embedding_vector_length, input_length=200) )
model.add(SpatialDropout1D(0.25))
model.add(LSTM(50, dropout=0.5, recurrent_dropout=0.5))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
history = model.fit(padded_sequence,sentiment_label[0],validation_split=0.2, epochs=10, batch_size=32)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')

plt.legend()
plt.show()
```

```python
def predict_sentiment(text):
    tw = tokenizer.texts_to_sequences([text])
    tw = pad_sequences(tw,maxlen=200)
    prediction = int(model.predict(tw).round().item())
    print("Predicted label: ", sentiment_label[1][prediction])


test_sentence1 = "I enjoyed my journey on this flight."
predict_sentiment(test_sentence1)

test_sentence2 = "This is the worst flight experience of my life!"
predict_sentiment(test_sentence2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report

# Sample data for demonstration
sentences = [
    "I love this product, it's fantastic!",
    "This movie was terrible, I hated it.",
    "The weather today is quite nice.",
    "The customer service was awful, very disappointing.",
    "I'm feeling happy and excited about the party tonight!"
]
labels = ['positive', 'negative', 'positive', 'negative', 'positive']

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Vectorize the sentences
X = vectorizer.fit_transform(sentences)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Train a linear support vector classifier
clf = LinearSVC()
clf.fit(X_train, y_train)

# Predict sentiment labels for the test set
y_pred = clf.predict(X_test)

# Evaluate the classifier
print(classification_report(y_test, y_pred))
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon
nltk.download('vader_lexicon')

def analyze_sentiment(text):
    # Initialize VADER
    sid = SentimentIntensityAnalyzer()
    # Analyze sentiment
    sentiment_scores = sid.polarity_scores(text)
    # Determine sentiment label
    if sentiment_scores['compound'] >= 0.05:
        return 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Example usage
text = "I love this product, it's amazing!"
sentiment = analyze_sentiment(text)
print("Sentiment:", sentiment)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download NLTK resources
nltk.download('vader_lexicon')

# Initialize the sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Define a function to perform sentiment analysis
def analyze_sentiment(text):
    """
    Analyzes the sentiment of the given text.
    Returns a dictionary with the scores for positive, negative, neutral, and compound sentiment.
    """
    scores = sid.polarity_scores(text)
    return scores

# Example text for sentiment analysis
text = "I love this product! It's amazing."

# Analyze the sentiment of the text
sentiment_scores = analyze_sentiment(text)

# Print the sentiment scores
print("Sentiment Scores:", sentiment_scores)

# Determine overall sentiment
if sentiment_scores['compound'] >= 0.05:
    print("Overall Sentiment: Positive")
elif sentiment_scores['compound'] <= -0.05:
    print("Overall Sentiment: Negative")
else:
    print("Overall Sentiment: Neutral")
```
